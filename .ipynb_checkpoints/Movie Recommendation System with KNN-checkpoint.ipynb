```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from scipy.sparse import csr_matrix
from fuzzywuzzy import process
from sklearn.neighbors import NearestNeighbors

model = NearestNeighbors(n_neighbors = 20,
                                 metric = 'cosine',
                                 algorithm = 'brute')
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# to read and merge my datasets

movie_df = pd.read_csv('Data/items.csv',usecols = ['movie_id','title'], sep='\t')
#print(movie_df)
#users_df = pd.read_csv('Data/users.csv',sep='\t')
#print(users_df)
rating_df = pd.read_csv('Data/ratings.csv',usecols = ['user_id','movie_id', 'rating'],sep='\t')
print(rating_df)
```

```python
# to merge the data into one dataframe
#user_rating_df = pd.merge(rating_df, users_df, on = 'user_id')
#print(user_rating_df)
df = pd.merge(rating_df, movie_df, on = 'movie_id')
df.head()
```

```python
# let me Create a user-movie matrix using pivot_table
# this will show wchich user liked a movie and how they rated the movies so that i can recommend it to a similar user

rating_info = df.pivot(index= 'movie_id', columns = 'user_id', values = 'rating')
rating_info.head()
```

```python
# There's NaN value which shows that the user probably have not watched the movie neither has the user rated the movie
# So, i will fill the NaN with zero's (0's) to show that the movie is not rated
rating_info = rating_info.fillna(0)
rating_info.head()
```

## Phase 3: Dimensionality Reduction

```python
#using sparse matrix for space efficiency for the non-zero values
matrix_reduc = csr_matrix(rating_info.values)
```

## Phase 4: Build the Recommendation Engine (KNN)

```python
model.fit(matrix_reduc)
```

```python
def recommender(movie_name, data, n):
    idx = process.extractOne(movie_name,movie_df['title'])[2] #using fuzzywuzzy[process,extractOne] for text matching for user input that might be incorrectly spelt
    print("Movie Selected :", movie_df['title'][idx], 'index : ', idx)
    print('Searching for recommendation ........')
    distance, indices = model.kneighbors(data[idx],n_neighbors = n) #to bring out how similar(distance) it is to the selected movie and where(indices) it's found on our movie_df
    similarities = (1-distance[0])*100 # To convert the similarity distance to %percentage for easy understanding
    recommended_movies = movie_df['title'].iloc[indices[0]]
    result_df = pd.DataFrame({'Recommended Movies' : recommended_movies.values,
                                'similarities(%)' : similarities.round(2)
                             }) #This shows the recommended movie in a tabular form
    result_df = result_df[result_df['Recommended Movies']!= movie_df['title'][idx]] # To exclude the selected movies from bieng recommended again
    return result_df
```

```python
def relationship(result_df):
    plt.figure(figsize = (8,5))
    sns.heatmap(
        result_df[['similarities(%)']].set_index(result_df['Recommended Movies']),
        annot=True, fmt=".1f", cmap="RdYlGn"
    )
    plt.title("RELATIONSHIP BETWEEN RECOMMENDED MOVIES AND SELECTED MOVIE")
    plt.ylabel("Recommeded Movieis")
    plt.tight_layout()
    plt.savefig("MOVIES RELATIONSHIP.png", dpi=300, bbox_inches='tight')
    plt.show()
    
```

### Testing the Recommendation system

```python
recommender("toy story", matrix_reduc, 5)
```

```python
results = recommender("toy story", matrix_reduc, 6)
relationship(results)
```
